```python
### ANN
```

```python
### Build the model
model = tf.models.Sequential()  ## blank model
### Add the layers
model.add(tf.layers.Flatten())  ## input Layer
model.add(tf.layers.Dense(784,activation="relu"))   ## hidden
model.add(tf.layers.Dense(10,activation="softmax")) ## output layer
### Compile the Model
model.compile(loss="sparse_categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python

### Scale the features
xtrain_scaled = xtrain/255
xtest_scaled = xtest/255
```

```python
### Train the model
history = model.fit(xtrain_scaled,ytrain,epochs=15,validation_data=(xtest_scaled,ytest))
```

```python
ypred = model.predict(xtest_scaled)
```

```python
model.evaluate(xtest_scaled,ytest)
```

```python
### Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest,np.argmax(ypred,axis=1))
cm
```

```python
### Accuracy
from sklearn.metrics import accuracy_score
a = accuracy_score(ytest,np.argmax(ypred,axis=1))
a
```

```python
plt.plot(history.history['accuracy'],label="Train Accuracy Curve")
plt.plot(history.history['val_accuracy'],label="Test Accuracy Curve")
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'],label="Train Loss Curve")
plt.plot(history.history['val_loss'],label="Test Loss Curve")
plt.legend()
plt.show()
```

```python
### Save the model
model.save('mnist.h5')
```
